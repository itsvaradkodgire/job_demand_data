## Accessing Data

```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
dataset=load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
```

```python
df.loc[:10,'salary_year_avg':'salary_hour_avg']
```

## Data Cleaning

```python
df0=df.copy()
```

```python
y_median=df0['salary_year_avg'].median()
```

```python
h_median=df0['salary_hour_avg'].median()
```

```python
df1=df.copy()
```

```python
df1['salary_year_avg']=df1['salary_year_avg'].fillna(y_median)
```

```python
df1['salary_hour_avg']=df1['salary_hour_avg'].fillna(h_median)
```

```python
df1
```

```python
df2=df1.copy()
```

```python
df2=df2.drop_duplicates()
```

```python
df2
```

```python
len(df1)-len(df2)
```

```python
df2=df2.drop_duplicates(subset=['job_title','company_name'])
```

```python
len(df2)
```

## Data Management

```python
df3=df.copy()
```

```python
df3.sample(10,random_state=42)
```

```python
df4=df.copy()
```

```python
df4.pivot_table(values='salary_year_avg',index='job_title_short',aggfunc='median')
```

```python
df4.groupby('job_title_short')['salary_year_avg'].median()
```

```python
df4.pivot_table(values='salary_year_avg',index='job_country',columns='job_title_short',aggfunc='median')
```

```python
top_countries=df4['job_country'].value_counts().head(6).index
```

```python
df_job_country_salary=df4.pivot_table(values='salary_year_avg',index='job_country',columns='job_title_short',aggfunc='median')
```

```python
df_job_country_salary=df_job_salary.loc[top_countries]
```

```python
job_titles=['Data Analyst','Data Engineer','Data Scientist']
```

```python
df_job_country_salary=df_job_country_salary[job_titles]
```

```python
df_job_country_salary.plot(kind='bar')
plt.ylabel('median salary $usd')
plt.xlabel('')
plt.title('salary by contry for job title')
plt.xticks(rotation=45,ha='right')
plt.show
```
